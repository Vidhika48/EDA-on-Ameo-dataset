```python
# Define outlier range
def get_outlier_range(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (lower_bound, upper_bound)

# Calculate outlier range for each gender group
lower_bound_male, upper_bound_male = get_outlier_range(df[df['Gender'] == 'm']['Salary'])
lower_bound_female, upper_bound_female = get_outlier_range(df[df['Gender'] == 'f']['Salary'])

outliers_male = df[(df['Gender'] == 'm') & ((df['Salary'] < lower_bound_male) | (df['Salary'] > upper_bound_male))]
outliers_female = df[(df['Gender'] == 'f') & ((df['Salary'] < lower_bound_female) | (df['Salary'] > upper_bound_female))]

sns.boxplot(data=df, x='Gender', y='Salary', showfliers=False)
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.title('Salary Distribution by Gender (without outliers)')
plt.show()

print("Number of outliers removed from male group:", outliers_male.shape[0])
print("Number of outliers removed from female group:", outliers_female.shape[0])
```

## Categorizing the designations

```python
substring_categories = {
    'engineer': 'Engineer',
    'manage': 'Manager',
    'analyst': 'Analyst',
    'developer': 'Developer',
    'consult': 'Consultant',
    'operat': 'Operations',
    'executive': 'Executive',
    'service': 'Service',
    'lecturer': 'Faculty',
    'faculty': 'Faculty',
    'admin': 'Admin'
}

# Function to categorize values based on substrings
def categorize_string(value):
    for substring, category in substring_categories.items():
        if substring in value:
            return category
    return 'Other'

# Create a new categorical column based on substring categories
df['Desg_cat'] = df['Designation'].apply(categorize_string)
```

# 4. No. of years/ time in the company

## Some values in DOJ are string values = present, since data was recorded in 2015 replacing present with last date of the year, and validating DOL > DOJ

```python
last_date= pd.Timestamp('2015-12-31')

df['DOL'] = df['DOL'].replace('present', last_date)

df.head()
```

```python
df['DOL'].dtypes
```

```python
# Convert 'DOJ' and 'DOL' columns to datetime
df['DOJ'] = pd.to_datetime(df['DOJ'])
df['DOL'] = pd.to_datetime(df['DOL'], errors='coerce')  # Convert 'present' to NaT

# Check if 'DOL' is not greater than 'DOJ' for each record
invalid_records = df[df['DOL'] < df['DOJ']]

print("Total number of invalid records:", len(invalid_records))

invalid_records.head()
```

```python
# Check if 'DOL' is greater than 'DOJ' for each record
valid_records_mask = df['DOL'] > df['DOJ']

# Create a new DataFrame containing only valid records
df_time = df[valid_records_mask].copy()

df_time.head()
```

```python
# Calculate the number of years in the company
df_time['Years_In_Company'] = (df_time['DOL'] - df_time['DOJ']).dt.days / 365.25  # Dividing by 365.25 for leap years

# Round the result to 2 decimal places
df_time['Years_In_Company'] = df_time['Years_In_Company'].round(2)

df_time['Years_In_Company'].head(10)
```

```python
avg_yrs = df_time['Years_In_Company'].mean()
max_yrs = df_time['Years_In_Company'].max()
min_yrs = df_time['Years_In_Company'].min()

print("avg yrs: ", round(avg_yrs, 2))
print("max yrs: ", round(max_yrs, 2))
print("min yrs: ", round(min_yrs, 2))
```

```python
# Plotting a histogram of years spent with the company
plt.figure(figsize=(10, 6))
plt.hist(df_time['Years_In_Company'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Years Spent with Company')
plt.ylabel('Frequency')
plt.title('Distribution of Years Spent with Company')
plt.grid(True)
plt.show()
```

```python
# Sort the DataFrame by the 'DOJ' column in ascending order
df_earliest_doj = df_time.sort_values(by='DOJ')

# Retrieve the first 5 records
top_5_earliest_doj = df_earliest_doj.head(3)

# Print the 5 records
print("5 records with the earliest Date of Joining (DOJ):")
print(top_5_earliest_doj)
```

```python
# Define outlier range 
def get_outlier_range(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (lower_bound, upper_bound)

# Calculate outlier range for 'Salary' column
lower_bound, upper_bound = get_outlier_range(df_time['Salary'])

# Remove outliers
df_no_outliers = df_time[(df_time['Salary'] >= lower_bound) & (df_time['Salary'] <= upper_bound)]

# Categorize the 'Years_In_Company' column
bin_edges = [0, 2, 5, 10, float('inf')]
bin_labels = ['0-2', '3-5', '6-10', '11+']
df_no_outliers['Years_Category'] = pd.cut(df_no_outliers['Years_In_Company'], bins=bin_edges, labels=bin_labels)

# Plot comparison
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_no_outliers, x='Years_Category', y='Salary')
plt.xlabel('Years in Company (Category)')
plt.ylabel('Salary')
plt.title('Years in Company vs. Salary (Without Outliers)')
plt.show()
```
